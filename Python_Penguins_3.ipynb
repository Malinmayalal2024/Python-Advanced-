# To fill null_values using fillna

```python
data["bill_length_mm"].fillna(data["bill_length_mm"].mean(),inplace=True)
```

```python
data["bill_depth_mm"].fillna(data["bill_depth_mm"].mean(),inplace=True)
```

```python
data["flipper_length_mm"].fillna(data["flipper_length_mm"].mean(),inplace=True)
```

```python
data["body_mass_g"].fillna(data["body_mass_g"].mean(),inplace=True)
```

```python
data["sex"].fillna(data["sex"].mode()[0],inplace=True)          # To change float values use mode()[0]
```

# Range function in pandas - Date_range

```python
pd.date_range(start="1997-03-18",end="2015-06-18")                      # To get all days
```

```python
pd.bdate_range(start="1997-03-18",end="2015-06-18")                     # To get only business dates [from monday to friday]
```
