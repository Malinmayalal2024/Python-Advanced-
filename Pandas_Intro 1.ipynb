# To fill or Relace null_values

```python
cd.isnull().sum()
```

```python
cd2=cd["bill_depth_mm"].mean();cd2
```

```python
cd["bill_depth_mm"]=cd["bill_depth_mm"].replace(np.nan,cd2)
```

```python
cd3=cd["flipper_length_mm"].mean();cd3
```

```python
cd["flipper_length_mm"]=cd["flipper_length_mm"].replace(np.nan,cd3)
```

```python
cd4=cd["body_mass_g"].mean();cd4
```

```python
cd["body_mass_g"]=cd["body_mass_g"].replace(np.nan,cd4)
```

```python
cd5=cd["sex"].mode().values[0];cd5                          # To change null value of float or string values[add .values]
```

```python
cd["sex"]=cd["sex"].replace(np.nan,cd5)
```
